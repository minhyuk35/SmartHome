```python
from pop import Cds
from pop import PwmController
import time

# ==============================
# PWM 초기화
# ==============================
pwm = PwmController()
pwm.init()
pwm.setChannel(-1)     # 모든 채널 초기화
pwm.setFreq(50)        # PWM 주파수 50Hz (기존과 동일)

cds = Cds(7)

THRESHOLD = 1500   
BRIGHT_CHANNEL = 2

last_state = None 

while True:
    val = cds.readAverage()
    print("CDS =", val)

    if val < THRESHOLD:
        if last_state != "BRIGHT":
            print("[AUTO] 어두움 감지 → 조명 켜짐")
            pwm.setChannel(BRIGHT_CHANNEL)

            for duty in range(0, 101, 10):
                pwm.setDuty(duty)
                time.sleep(0.1)

            last_state = "BRIGHT"

    else:
        if last_state != "OFF":
            print("[AUTO] 밝음 감지 → 조명 꺼짐")
            pwm.setChannel(BRIGHT_CHANNEL)
            pwm.setDuty(0)    # LED OFF
            last_state = "OFF"

    time.sleep(0.1)
```
